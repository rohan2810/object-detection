```python
import cv2
import os
count=1
vid = "irobot/videos/trim_36.mp4"
vidcap = cv2.VideoCapture(vid)
def getFramePerSecond(sec):
    vidcap.set(cv2.CAP_PROP_POS_MSEC,sec*1000)
    results,image = vidcap.read()
    if results:
        cv2.imwrite("irobot/frames/trim_36/"+str(count)+".jpg", image)
    return results
sec = 0
frame_rate = 0.155
    
success = getFramePerSecond(sec)
while success:
    count = count + 1
    sec = sec + frame_rate
    sec = round(sec, 2)
    success = getFramePerSecond(sec)
```
